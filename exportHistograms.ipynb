# Export Histograms
This script will parse a given folder containing data saved from Imaris and plot a histogram of the shortest distances to the surface.

```python
import csv
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
dataFolder = ('\\\\pn.vai.org\\projects\\patel\\vari-core-generated-data\\Arlene\\Mito and hmg-5 images to analyze\\Measurements')

# Get list of subfolders
subfolders = []
for item in os.listdir(dataFolder):
    item_path = os.path.join(dataFolder, item)
    if os.path.isdir(item_path):
        subfolders.append(item)

print(subfolders)
```

```python
binWidth = 0.1

for folder in subfolders:
    print(folder)
    filepath = os.path.join(dataFolder, folder, 'mtDNA_Statistics', 'mtDNA_Shortest_Distance_to_Surfaces_Surfaces=Mitochondria.csv')
    
    mtDistances = []
    
    with open(filepath, 'r', newline='') as csvfile:
        reader = csv.reader(csvfile)
    
        #Skip the first three rows
        for _ in range(4):
            next(reader)
        
        for row in reader:
            if row:
                mtDistances.append(float(row[0]))
                #print(row[0])

        plt.hist(mtDistances, bins=np.arange(-1, 5, binWidth) , edgecolor='black')
        plt.xlabel(r'Distances ($\mu$m)')
        plt.ylabel('Num mtDNA spots')
        
        plt.savefig(os.path.join(dataFolder, folder + '.png'))
        plt.clf()
```
